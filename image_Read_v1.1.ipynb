```python
# Contents for your test_ocr.py file

import pytesseract
from PIL import Image
import platform

print(f"Python Architecture: {platform.architecture()}")
print("-" * 30)

try:
    # 1. Set the path to Tesseract
    pytesseract.pytesseract.tesseract_cmd = r'C:\Tesseract-OCR\tesseract.exe'
    print("Tesseract path set to: C:\\Tesseract-OCR\\tesseract.exe")

    # 2. Set the path to your image
    image_path = r'C:\Users\Rahul Sharma\OneDrive - UWA\UWA\Sem_3_Capstone Project\image_3_172.jpeg'
    print(f"Attempting to open image: {image_path}")
    
    img = Image.open(image_path)
    print("Image opened successfully.")

    # 3. Run OCR
    print("Calling Tesseract...")
    text = pytesseract.image_to_string(img)

    print("\n--- OCR SUCCESS! ---")
    print("Tesseract was executed correctly.")
    print("\n--- Recognized Text ---")
    print(text)

except pytesseract.TesseractNotFoundError:
    print("\n!!! TESSERACT NOT FOUND ERROR !!!")
    print("Even with the correct path, Python could not launch tesseract.exe.")
    print("This strongly suggests an environment or permission issue.")

except FileNotFoundError:
     print(f"\n!!! IMAGE NOT FOUND ERROR !!!")
     print(f"Could not find the image file at: {image_path}")

except Exception as e:
    print(f"\n!!! AN UNEXPECTED ERROR OCCURRED !!!")
    print(e)
```
